```python
# Impor library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Baca data dari file CSV menjadi DataFrame
df = pd.read_csv('/workspaces/147965930/project/final/sql/tahun.csv')

# Mendefinisikan palet warna yang berbeda untuk setiap tahun
palette = sns.color_palette("husl", len(df))

# Setel gaya plot menjadi background abu-abu
sns.set_style("darkgrid")

# Contoh visualisasi menggunakan seaborn dengan palet warna yang berbeda untuk setiap tahun
ax = sns.barplot(x='Tahun', y='Jumlah', data=df, palette=palette)
plt.title('Jumlah Gempa Pertahun', fontsize=16, fontweight='bold', color='blue')
plt.ylabel('Jumlah', fontsize=14, fontweight='bold', color='green')

# Tambahkan teks sejajar di atas grafik
for index, value in enumerate(df['Jumlah']):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=8, fontweight='bold', color='black', alpha=0.8)

# Setel label pada sumbu x agar sejajar dengan grafik
ax.set_xticklabels(df['Tahun'], rotation=90, fontsize=8, fontweight='bold', color='purple')

plt.show()
```
